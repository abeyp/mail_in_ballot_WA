```python
all_years = df.columns[1:]
adams_perc = df.loc[0][1:].replace('-',np.nan).astype(float)

#TODO: make wider, set yaxis to 1

plt.plot(all_years,adams_perc, linestyle='-', marker='o')
plt.plot(all_years, df.loc[1][1:].replace('-',np.nan).astype(float))

plt.title('Ballot Drop Box Usage by Year')
plt.xlabel('Year')
plt.ylabel('% of total ballots returned')
plt.xticks(rotation=45)
plt.show()
```

```python

# Generate some sample data
years = [year for year in range(2012, 2024)]

#King county percentage of total ballots rrerturned 
values = [0.21,0.20,0.22,0.26,0.49,0.53,0.40,0.47,0.74,0.57,0.60,0.55]
data = {'date_column': years, 'value_column': values}
test_df = pd.DataFrame(data)

# Create the plot
plt.plot(test_df['date_column'], test_df['value_column'], linestyle = 'dotted')

# Add title and axis labels
plt.title('Time Series of Total Ballots Returned in King County')
plt.xlabel('Time')
plt.ylabel('Value')
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Grab rows for King, Snohomish, Pierce

county_df = df.loc[df["County"].isin(["King", "Snohomish","Pierce"])]
county_df
```

```python
county_df.set_index('County', inplace=True)
```

```python
county_df.dtypes
```

```python
county_df = county_df.astype('float64')
county_df
```

```python
plt.plot(county_df.loc['King'])
```

```python
county_df.index
```

```python
plt.rcParams["font.size"] = 12
plt.rcParams["figure.figsize"] = (18, 10)

for x in county_df.index:
    plt.plot(county_df.loc[f"{x}"], label=x)

plt.title('Ballots Returned of Largest Counties')
plt.xlabel('Year')
plt.ylabel('% of Ballots Returned')
plt.xticks(rotation=45)

plt.legend()
plt.show()
    
```
