## On and Off Years for Drop Box Usage

```python
# Plot on and off years for Drop Box Usage

x_statewide = [0.367, 0.388, 0.405, 0.407, 0.569, 0.546, 0.450, 0.483, 0.731, 0.563, 0.611, 0.560]
y_years = df.columns.astype(int)
print(x_statewide, y_years)

statewide = {'year': y_years, 'percentage':x_statewide}
statewide_df = pd.DataFrame(data=statewide)

statewide_df
```

```python
#create on_years and off year variables

on_years = []
off_years = []
for x in y_years:
    if x % 4 == 0: 
        on_years.append(x)
    else: 
        off_years.append(x)
        
on_years.sort()
off_years.sort()

print(on_years)
print(off_years)
```

```python
on_years_df = statewide_df.loc[statewide_df['year'].isin(on_years)]
off_year_df = statewide_df.loc[statewide_df['year'].isin(off_years)]
```

```python
#On Years
plt.plot(on_years_df['year'], on_years_df['percentage'], 
         label='Presidential General Elections',color = 'orange', marker = 'o')

#Off Years
plt.plot(off_year_df['year'], off_year_df['percentage'], 
         label='Off Years',color='#1f77b4', marker = 'o')


plt.title('Drop Box Usage during On and Off Years')
plt.ylabel('Usage (%)')

plt.ylim([0,1])
plt.xticks(on_years)

# vals = ax.get_yticks()
# ax.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
    
plt.grid()
plt.legend()
plt.show()
```
